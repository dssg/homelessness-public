```python
import clean, features, models, pipeline
```

```python
m = clean.load('master')
```

```python
feature_sets = {'All': features.all_sets,
                'Reduced': features.reduced}

targets = ['CaseSuccess',
           'Reentered12MonthFromPermanent']

classifiers = {'logistic': models.logistic,
               'random_forest_large': models.random_forest_large,
               'decision_tree': models.decision_tree}
```

```python
p = {}
```

```python
p['Master'] = pipeline.Pipeline(m, 'master', feature_sets, targets, classifiers)
# comment out below if you don't want to re-run the models
# p['Master'].model()
```

```python
for pt, path in {'Services Only program (HUD)':        'sso',
                 'Permanent supportive housing (HUD)': 'psh',
                 'Transitional housing (HUD)':         'th',
                 'Rapid Re-Housing (HUD)':             'rrh',
                 'Emergency Shelter (HUD)':            'es'}.iteritems():
    p[pt] = pipeline.Pipeline(m[m['ProgramTypeAggregate'] == pt], path, feature_sets, targets, classifiers)
    # comment out below if you don't want to re-run the models
    # p[pt].model()
```

# Results

```python
plt.figure(figsize=(12, 12))
p['Master'].plot_roc(feature_sets=['All'], targets=['CaseSuccess'])
p['Master'].plot_roc(feature_sets=['Reduced'], targets=['CaseSuccess'], classifiers=['logistic'])
```

```python
plt.figure(figsize=(12, 12))
p['Master'].plot_pr(feature_sets=['All'], targets=['CaseSuccess'])
p['Master'].plot_pr(feature_sets=['Reduced'], targets=['CaseSuccess'], classifiers=['logistic'])
```

```python
plt.figure(figsize=(12, 12))
for pt, pipeline in p.iteritems():
    pipeline.plot_roc(feature_sets=['All'], targets=['CaseSuccess'], classifiers=['random_forest_large'], label_prefix=pt)
```

```python
plt.figure(figsize=(12, 12))
for pt, pipeline in p.iteritems():
    pipeline.plot_pr(feature_sets=['All'], targets=['CaseSuccess'], classifiers=['random_forest_large'], label_prefix=pt)
```
